```python
import numpy as np

t = np.linspace(0.5,2.8,1000)
x = 16 * np.sin(t) ** 3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - 4 * np.cos(4*t)
# help(np.clip)
# yd = np.clip(y,0,None)
xd = np.clip(x,0,None)

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10,10))
# ax = fig.add_subplot(1, 1, 1) # nrows, ncols, index
# ax.set_facecolor('yellow')
# plt.subplot(1,2,1)
# plt.plot(xd,y)
# plt.plot(-xd,y)
ax = plt.gca()
ax.set_facecolor('black')
ax.axes.xaxis.set_visible(False)
ax.axes.yaxis.set_visible(False)
mag = np.sqrt(x**2 + y**2)
xn = x/mag
yn = y/mag
d = 0.5
w = 2
switch_bool = False
c1 = 'red'
c2 = 'blue'
for f in [1 + d * i for i in range(0,6)]:
    plt.plot( f * xd,f * y, c = c1, linewidth = w)
    plt.plot(-f * xd,f * y, c = c2, linewidth = w)
    # if f - 1 >= d:
    if switch_bool:
        plt.plot([ f * xd[-1],  (f-d) * xd[-1]],[f *y[-1], (f-d) * y[-1]],c=c1, linewidth = w)
        plt.plot([-f * xd[-1], -(f-d) * xd[-1]],[f *y[-1], (f-d) * y[-1]],c=c2, linewidth = w)
    else:
        plt.plot([ f * xd[0],  (f-d) * xd[0]],[f *y[0], (f-d) * y[0]],c=c1, linewidth = w)
        plt.plot([-f * xd[0], -(f-d) * xd[0]],[f *y[0], (f-d) * y[0]],c=c2, linewidth = w)
    switch_bool = not switch_bool

# plt.axis('off')


# plt.subplot(1,2,2)
# plt.scatter(t,t,c=t)
```
